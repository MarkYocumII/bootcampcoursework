#### Import Dependencies

[Documentation for `random` module](https://docs.python.org/2/library/random.html)

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

#### Create data

```python
# The maximum x value for our chart will be 100
x_limit = 100

# List of values from 0 to 100 each value being 1 greater than the last
x_axis = np.arange(0, x_limit, 1)

# Create a random array of data that we will use for our y values
data = [random.random() for value in x_axis]
```

#### Create a scatter plot
- The size of each point on our plot is determined by their x value


##### Arguments:
- **x** -> data for x-axis
- **y** -> data for y-axis
- **marker**-> shape to use for datapoints
- **facecolors** -> color of the center of each datapoint
- **edgecolors** -> color of the edge of each datapoint
- **s** -> data to determine the size of each datapoint
- **alpha** -> transparenct of the datapoints (0 = completely transparent)

```python
plt.scatter(x=x_axis, y=data, marker="o", facecolors="orange", edgecolors="blue", s=x_axis, alpha=0.8)
```

#### The y limits of our scatter plot is 0 to 1

```python
plt.ylim(0, 1)
```

#### The x limits of our scatter plot is 0 to 100

```python
plt.xlim(0, x_limit)
```

#### Prints the scatter plot to the screen

```python
plt.show()
```
