#### Import dependencies

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Create the X and Y axis lists using the data from the instructions

- **HINT:** your x-axis is going to be the time of the rollercoaster.
- **HINT:** your y-axis data is provided in the instructions

```python
time = np.arange(0,130,10)
DD_speeds = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
RG_speeds = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

#### Plot the charts and apply some styling

```python
danger_drop, = plt.plot(time, DD_speeds, color="red", label="Danger Drop")
railgun, = plt.plot(time, RG_speeds, color="blue", label="RailGun")
```

#### Add labels to X and Y axes and also add title

```python
plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
```

#### Set the limits for the X and Y axes

```python
plt.xlim(0,120)
plt.ylim(5,100)
```
